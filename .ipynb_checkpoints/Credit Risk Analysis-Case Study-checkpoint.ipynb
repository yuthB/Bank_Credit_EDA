## Analysing the dataset based on income and credit

```python
# Dividing the income in different bins
bins = [0,25000,50000,75000,100000,125000,150000,175000,200000,225000,250000,275000,300000,325000,350000,375000,400000,425000,450000,475000,500000,10000000000]
slot = ['0-25000', '25000-50000','50000-75000','75000-100000','100000-125000', '125000-150000', '150000-175000','175000-200000',
       '200000-225000','225000-250000','250000-275000','275000-300000','300000-325000','325000-350000','350000-375000',
       '375000-400000','400000-425000','425000-450000','450000-475000','475000-500000','500000 and above']

df['AMT_INCOME_BIN']=pd.cut(df['AMT_INCOME_TOTAL'],bins,labels=slot)
```

`Inferences:`
1. Female Applicants are significantly higher than Male Applicants.
2. People having Income in the range of 25000 to 2 Lakhs have applied more for the loan.
3. In the same Category of 25,000 to 2  Lakhs income Female Applicants are much higher than Male Applicants.
4. Number of Applicants with income above 4 Lakh is very less.

```python
# Dividing the credit amount in different bins
bins = [0,150000,200000,250000,300000,350000,400000,450000,500000,550000,600000,650000,700000,750000,800000,850000,900000,1000000000]
slots = ['0-150000', '150000-200000','200000-250000', '250000-300000', '300000-350000', '350000-400000','400000-450000',
        '450000-500000','500000-550000','550000-600000','600000-650000','650000-700000','700000-750000','750000-800000',
        '800000-850000','850000-900000','900000 and above']

df['AMT_CREDIT_BIN']=pd.cut(df['AMT_CREDIT'],bins=bins,labels=slots)
```

`Inferences:`
1. Females have applied for a higher credit than Males.
2. Applicants both Male and Female have applied for credit 9 Lakhs and above which seems as an outlier

## Finding the Imbalance in data

```python
# Separating all the target data to find imbalance
target_0=df[df['TARGET']==0]
target_1=df[df['TARGET']==1]
```

```python
# Calculating imbalance ratio
print('Imbalance ratio is ',round(len(target_0)/len(target_1),2))
```

## Looking at the ratio: Majority target is 0 and Minority target is 1

```python
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
sns.countplot(x=target_1['AMT_INCOME_BIN'],hue=target_1['CODE_GENDER'],palette='rocket')
plt.title("Distribution of Applicant's Income Target 1")
plt.xlabel("Income Categories")
plt.ylabel("Number of Applicants")
plt.legend(title='Gender')
plt.xticks(rotation=45)

plt.subplot(1,2,2)
sns.countplot(x=target_0['AMT_INCOME_BIN'],hue=target_0['CODE_GENDER'],hue_order=['M','F'],palette='rocket')
plt.title("Distribution of Applicant's Income Target 0")
plt.ylabel("Number of Applicants")
plt.legend(title='Gender')
plt.xticks(rotation=45)

plt.show()
```

`Inferences:` Most of the Applicants are in the Income range of 50k to 2.25 Lakh in both Male and female Category.


```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.countplot(x=target_1['CODE_GENDER'])
plt.title('Target 1 Applicant Distribution')
plt.xlabel('Gender')
plt.subplot(1,2,2)
sns.countplot(x=target_0['CODE_GENDER'],order=['M','F'])
plt.title('Target 0 Applicant Distribution')
plt.xlabel('Gender')
plt.show()
```

`Inference:` Clearly Male applicants are likely to be at default than their female counterparts.

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.boxplot(data=target_1,x='CODE_GENDER',y='AMT_INCOME_TOTAL')
plt.subplot(1,2,2)
sns.boxplot(data=target_0,x='CODE_GENDER',y='AMT_INCOME_TOTAL',order=['M','F'])
```

#### Checking the outlier  in Target_1 and Target_0 dataset

```python
target_1=target_1[target_1['AMT_INCOME_TOTAL']<=100000000]
target_0=target_0[target_0['AMT_INCOME_TOTAL']<=10000000]
```

`Inference:` We can remove this outlier as it is affecting rest of the analysis

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=target_1,x='CODE_GENDER',y='AMT_INCOME_TOTAL')
plt.title('Income Distribution in Target 1 (Defaulters)')
plt.xlabel('Gender')
plt.ylabel('Income')
plt.subplot(1,2,2)
sns.boxplot(data=target_0,x='CODE_GENDER',y='AMT_INCOME_TOTAL',order=['M','F'])
plt.title('Income Distribution in Target 0 (Non-Defaulters)')
plt.xlabel('Gender')
plt.ylabel('Income')
plt.show()
```

`Inferences:`
1. People with higher income tend are likely to pay the loans on time.
